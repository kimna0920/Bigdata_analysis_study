# 2. 연산자 사용하기기

```python
#산술 연산자
print(10**2) #100
print(10*2) #20
print(10%2) #0
print(5/2)  #2.5
print(5//2) #2
```

```python
#비교연산자 (참:True, 거짓:False)
print(10>=2)  #Truet
print(10<=2)  #False
print(10==2)  #Falsef
print(10 !=2) #True
print(10%2 ==1) #False
```

```python
#논리연산자(참:True, 거짓:False)
#and(둘 다 참일때 True), or (둘중하나만 참일때 True), not(반전) 연산자
print(1==1 and 2 !=1)       #True
print(10%2 !=0 and 1+1 >0)  #False
print()
#and 연산자
print(1==1 and 2 != 1)      #True
print(10%2 !=0 and 1+1 >0)  #False
print()

#or 연산자
print(10<5 or 10==5)        #False
print(10%2 !=0 or 1+1 >0)   #True
print()         

#not 연산자
print(not 10>5)             #False
print(not 10==5)            #True
print(not 0)                #True
print(not 10)               #False
```

```python
#비밀번호 확인
passwd = int(input('비밀번호 4자리를 입력하세요:'))

if passwd == 2020:
  print('비밀번호가 일치 합니다')
```

```python
#비밀번호 해킹 프로그램
for i in range(1,100000):
    if i == 2020:
      print('비밀번호가 일치합니다')
      # print(f'비밀번호는 {i}입니다')
      print('비밀번호는 '+str(i)+'입니다')
```

```python
#else 문 추가
passwd = int(input('비밀번호 4자리를 입력하세요:'))

if passwd == 2020:
  print('비밀번호가 일치합니다.')
else:
  print('비밀번호가 일치하지 않습니다.')
```

```python
#심리게임 프로그램
print('[소름끼치도록 놀라운 심리테스트]')

menu = input('당신이 좋아하는 음식을 입력해주세요:')

# print('당신은'+menu+'을 좋아하는 사람입니다.')

if menu=='짜장면':
  print('당신은 짜장면을 좋아하는 사람입니다.')
elif menu=='아이스크림':
  print('당신은 아이스크림을 좋아하는 사람입니다.')
elif menu=='사탕':
  print('당신은 사탕을 좋아하는 사람입니다.')
else:
  print('당신은 짜장면, 아이스크림, 사탕을 좋아하지 않는 사람입니다.')
```
