```python
#다음은 2010~2019년 기온 데이터(년도, 평균, 최소, 최대)이다.
# - 홀수년도 만 출력
# - 년도와 평균기온 딕셔너리 만들기
# - 평균기온 최소일 때 해당년도 추출하기 예) 2012년 - 14.5도
# - 평균기온 최대일 때 해당년도 추출하기 예) 2016년 - 19.6도
# - 최저기온 최소일 때 해당년도 추출하기 예) 2012년 - 18.6도

y = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
yavg = [14.9, 14.6, 14.5, 15.3, 15.1, 15.7, 15.2, 15.1, 15.7, 16.2]
ymax = [34.1, 33, 34.5, 35, 32.9, 33.5, 37.3, 36.2, 36.4, 35]
ymin = [19, 18.8, 18.6, 19.5, 19.2, 19.5, 19.8, 19.6, 19.2, 19.8]

# 홀수년도
yodd = [year for year in y if year % 2 == 1]
print(yodd)

# 년도와 평균기온 딕셔너리 만들기
ytemp = dict(zip(y, yavg))
print(ytemp)

# 평균기온 최소일 때 해당년도 추출하기
mintemp = min(yavg)
ryear = [year for year, temp in ytemp.items() if temp == mintemp]
print(f"평균기온 최소 : {ryear[0]}년 - {mintemp}도")

# 평균기온 최대일 때 해당년도 추출하기
maxtemp = max(yavg)
ryear = [year for year, temp in ytemp.items() if temp == maxtemp]
print(f"평균기온 최대 : {ryear[0]}년 - {maxtemp}도")

# 최저기온 최소일 때 해당년도 추출하기
minmintemp = min(ymin)
ryear = [year for year, temp in zip(y, ymin) if temp == minmintemp]
print(f"최저기온 최소 : {ryear[0]}년 - {minmintemp}도")
```

```python
filename = '/content/drive/MyDrive/빅데이터/부산.csv'

import csv
f = open(filename,'r',encoding='cp949')#윈도우 한글 인코딩 방식식
data = csv.reader(f,delimiter=',')
header = next(data)
for row in data:
  print(row[2])
f.close()
```

```python
#부산의 기온이 가장 높았던 날의 날짜와 기온 구하기
filename = '/content/drive/MyDrive/빅데이터/부산.csv'

import csv
with open(filename, 'r', encoding='cp949') as f:
    data = csv.reader(f, delimiter=',')
    header = next(data)
    max_temp = -999
    max_date = ''
    for row in data:
        if row[2] == '':
            row[2] = -999
        else:
            row[2] = float(row[2])
        if max_temp < row[2]:
            max_date = row[0]
            max_temp = row[2]

print(f'부산최대 온도는 {max_date}에 {max_temp}도')
```

```python
# 부산의 기온이 가장 낮았던 날의 날짜와 기온 구하기
filename = '/content/drive/MyDrive/빅데이터/부산.csv'

import csv

with open(filename, 'r', encoding='cp949') as f:
    data = csv.reader(f)
    header = next(data)
    min_temp = float('inf')
    min_date = ''
    for row in data:
        if row[2] == '':
            row[2] = float('inf')
        else:
            row[2] = float(row[2])
        if min_temp > row[2]:
            min_date = row[0]
            min_temp = row[2]

print(f"부산최저 온도는 {min_date}에 {min_temp}도.")
```
