  * len(): 문자열 길이 구하기
  * count(): 문자의 개수 구하기
  * find(): 문자가 처음 나오는 위치 반환, 없으면 -1
  * index(): 문자열이 처음 나오는 위치 반환, 없으면 오류
  * join(): 문자열 사이에 문자 삽입, 예: '삽입문자'.join(문자열)
  * upper(): 소문자를 대문자로 변환
  * lower(): 대문자를 소문자로 변환
  * replace(): 문자열 바꾸기, 예:replace(원본, 변환 문자)
  * split(): 문자열 나누기 - 인수가 없으면 공백(스페이스, 탭, 엔터 등 구분자)을 기준

# 3)딕셔너리

*   특징

  *   키(key)와 값(value)쌍을 요소로 가짐
  *   중괄호로 작성되며 각 요소는 쉼표로 구분
  *   순서 없음
  *   키는 변경할 수 없으며 값은 변경가능
*   딕셔너리 접근
  *   특정 요소 값 : 딕셔너리명[키]
  *   요소 키 커렉션 : 딕셔너리명.keys()
  *   요소 값 컬렉션 : 딕셔너리명.values()
  *   키와 여소 컬렉션은 list() 생선자로 리스트로 변환가능


*   요소 추가
  * 딕셔너리명[추가키]=값

* 요소 수정
  * 딕셔너리명[키] = 값
* 요소 삭제
  * del딕셔너리명[키]
  * 딕셔너리명.pop(키)


```python
#딕셔너리
dt={'a':200,'b':150,'c':100}
print(dt)
```

```python
#딕셔너리 키
dt={'a':200,'b':150,'c':100}
key = list(dt.Keys())
print(key)
```

```python
#딕셔너리 값
dt={'a':200,'b':150,'c':100}
print((dt.values()))
```

```python
#딕셔너리 출력
dt={'a':200,'b':150,'c':100}
print(dt['a'])
print(dt['b'])
print(dt['c'])
```

```python
#딕셔너리 추가
dt={'a':200,'b':150,'c':100}
dt['d']=300
print(dt)
dt['b']=400
print(dt)
dt.update({'e':500})
print(dt)
```

```python
#딕셔너리 삭제
dt={'a':200,'b':150,'c':100}
del dt['b']
print(dt)
dt.pop('a')
print(dt)
#모두 지우기
dt.clear()
print(dt)
```

# 4)집합


*   특징
  *   중복 없는 요소로만 구성
  *   중괄호로 작성되며 각 요소는
  *   순서없음음


*   생성자 : set()
  * 리스트나 튜플의 중복 요소 제거
*요소 추가
  * 집합명.add(요소) : 하나의 요소 추가
  * 집합명.update(추가집합) : 여러 요소를 추가
* 요소 삭제
  * 집합명.remove(요소값) : 하나의 요소 삭제
  * 집합명.clear() : 모든 요소 삭제
* 집합 연산
  * 교집합 : &
  * 합집합 : |
  * 차집합 : ~


```python
#집합
a=(1,2,3,4) #튜플
b=[10,20,30,30] #리스트
a = set(a) #중복 제거
b = set(b)

print(a)
print(b)

a.add(100)
b.update({100,200,300})
print(a)
print(b)
```
